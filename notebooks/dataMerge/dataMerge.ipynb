```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv("data/index.csv", index_col=0)
print("Original column names:")
print(df1.columns)
print(df1.shape)
df1.columns = ['zipcode', 'renovation_index', 'renovation_index_with_rent']
df1['zipcode'] = df1['zipcode'].astype(str)
df1.head()
```

```python
df2 = pd.read_csv("data/crime.csv")
print("Original column names:")
print(df2.columns)
print(df2.shape)
df2.drop(['ZIPCODE.1'], axis=1, inplace=True)
df2.columns = ['zipcode', 'crime_change_abs', 'crime_change_rate']
df2['zipcode'] = df2['zipcode'].astype(str)
df2.head()
```

```python
df3 = pd.read_csv("data/complaintFinal.csv", index_col=0)
print("Original column names:")
print(df3.columns)
print(df3.shape)
df3.columns = ['zipcode', 
               'SS_change_rate', 'H_change_rate', 'N_change_rate', 'T_change_rate', 'PHS_change_rate', 'VP_change_rate',
               'SS_change_pop', 'H_change_pop', 'N_change_pop', 'T_change_pop', 'PHS_change_pop', 'VP_change_pop',
               'SS_change_area', 'H_change_area', 'N_change_area', 'T_change_area', 'PHS_change_area', 'VP_change_area']
df3['zipcode'] = df3['zipcode'].astype(str)
df3.head()
```

```python
df4 = pd.read_csv("data/mastercardFinal.csv", index_col=0)
print("Original column names:")
print(df4.columns)
print(df4.shape)
df4.columns = ['zipcode', 'MasterCard_growth_rank']
df4['zipcode'] = df4['zipcode'].astype(str)
df4.head()
```

```python
df5 = pd.read_csv("data/privateVehicle_subway_change.csv", index_col=0)
print("Original column names:")
print(df5.columns)
print(df5.shape)
df5.columns = ['zipcode', 'subway_change_abs', 'subway_change_rate', 'yellowTaxi_change_abs',
               'yellowTaxi_change_rate', 'privateVehicle_change_abs']
df5['zipcode'] = df5['zipcode'].astype(str)
df5.head()
```

### Merge

```python
df = pd.merge(df1, df2, on='zipcode')
df = pd.merge(df, df3, on='zipcode')
df = pd.merge(df, df4, on='zipcode')
df = pd.merge(df, df5, on='zipcode')
df.head()
```

```python
df.shape
```

```python
df.to_csv("output/dataInnerMerge.csv")
```

```python
df = pd.merge(df1, df2, on='zipcode', how='outer')
df = pd.merge(df, df3, on='zipcode', how='outer')
df = pd.merge(df, df4, on='zipcode', how='outer')
df = pd.merge(df, df5, on='zipcode', how='outer')
df.head()
```

```python
df.shape
```

```python
df.to_csv("output/dataOuterMerge.csv")
```

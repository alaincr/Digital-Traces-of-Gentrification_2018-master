```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
import geopandas as gpd
from shapely.geometry import Point
import fiona
```

```python
df = pd.read_csv("monthlydata/complaintTimeforAll.csv", index_col=0)
df.head()
```

```python
df_ = df.T
df_.head()
```

```python
df_roll = df_.iloc[2:,:].rolling(12).mean()
df_roll.head(15)
```

```python
df_temp1 = df_.iloc[:2,:]
df_temp2 = df_roll.iloc[11:,:]
```

```python
df_new = df_temp1.append(df_temp2)
df_new.head()
```

```python
dftime = df_new.T
dftime.head()
```

```python
dftime['type'].unique()
```

```python
import re
import string
```

```python
temp = [re.sub('['+string.punctuation+']', " ", onetype) for onetype in dftime['type'].unique()]
namelist = ["_".join(one.split()) for one in temp]
namelist
```

```python
for i,name in enumerate(dftime['type'].unique()):
    print(name)
    df_temp = dftime[dftime['type'] == name]
    print(df_temp.shape)
    df_new = df_temp.drop(columns=['type']).fillna(0.5)
    df_new.to_csv("complaint/" + namelist[i] + ".csv")
    print(namelist[i] + ".csv done...")
```
